```python
# We will start with importing necessary packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Now we will load dataset using the link.
europe = pd.read_csv('https://raw.githubusercontent.com/arpgpt/data-visualization/master/gdp_europe.csv', index_col='country')
```

```python
europe.head()
```

```python
#our first job is to convert the column name from string to year integer. We will use strip() method.
years = europe.columns.str.strip('gdpPercap_')
europe.columns = years.astype(int)
europe.head()
```

```python
#First we will draw very simple 
europe.plot()
plt.title('GDP Per Capita')
```

```python
#Since this graph doesn't give us appropriate visualisation. Therefor we will draw another type of graph.
europe.T.plot()
plt.title("GDP Per Capita")
```

```python
#Since all the lines at once on the plot is not readable, we will try to draw them seperately.
#Here is how we draw single line graph.
europe.loc['Albania'].plot()
plt.title('GDP Per Capita')
```

# We can also customise the line using different properties such as:
* linestyle: '-','--','-.',':'
* linewidth : int or float
* color : 'any color name'
* alpha : any value between 0 and 1
*<p> And a method annotate(string,position,color)</p>

```python
europe.loc['Albania'].plot(color='black', linewidth='4',alpha=0.9, linestyle=':')
europe.loc['Poland'].plot(color='red', linewidth='1', alpha=0.4, linestyle='-')
plt.annotate('GDP Albania',xy=(1980,9300),color='black')
plt.annotate('GDP Poland',xy=(1990,2000),color='red')
plt.title('GDP Per Capita')
```

```python
#We can also plot graph like indicating every label and GDP coordinate along with line.
#Here is how we do it.
df = europe.loc['Albania']
xs = list(years)
ys = [int(df.iloc[i]) for i in range(len(df))]
plt.plot(xs,ys,'bo-')
for x,y in zip(xs,ys):
    label = x,y,
    plt.annotate(label, #This is the string that will be written.
                 (x,y), #This is the position where string will be written
                 textcoords='offset points', #How to position the text.
                 xytext=(0,0), #Distance from the centre of point (x,y)
                 ha='left' #Horizontal Alignment can be set to 'left','right','centre'
                )
```
