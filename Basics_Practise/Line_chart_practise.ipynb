We will learn about a visualization tool namely **Line Chart**, and learn how to create it using Matplotlib. 
    
    
A **line chart** or line graph is a type of chart which displays information as a series of data points called ‘markers’ connected by straight line segments. It is a basic type of  chart common in many fields. 

A line chart is often used to visualize a trend in data over intervals of time – a time series – thus the line is often drawn chronologically. In these cases they are known as run charts.

## The Dataset: gdp of Countries in North America from year 1952 to 2007
The dataset contains the gdp data for the countries in North America. The current version presents data pertaining to 14 countries.
Data can be found [here](https://raw.githubusercontent.com/arpgpt/data-visualization/master/north_america.csv)


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

We will now read our data set.

```python
NorthA = pd.read_csv("https://raw.githubusercontent.com/arpgpt/data-visualization/master/north_america.csv", index_col = 'country')
NorthA.head()
```

- Before plotting, we convert the column headings from a string to integer data type, since they represent numerical values.
- Extract year from last 4 characters of each column name.
- The current column names are structured as 'gdpPercap_(year)', 
- so we want to keep the (year) part only for clarity when plotting GDP vs. years
- To do this we use strip(), which removes from the string the characters stated in the argument
- This method works on strings, so we call str before strip()

```python
# First fetch the values of headers
years = NorthA.columns.str.strip('gdpPercap_')
NorthA.columns = years.astype(int)
#NorthA.columns = NorthA.columns.str.strip('gdpPercap_').astype(int)

#Here I don't know why do i have to run from the begining to work this segment of code.
#Otherwise it gives this error: AttributeError: Can only use .str accessor with string values!
```

We will now plot a line chart for all the countries in North America

```python
NorthA.plot()
plt.ylabel("GDP Per Capita")
```

This plot looks confusing and crowded. Let us clean the data a bit more and plot a line chart after transposing the axis.

```python
NorthA.T.plot()
plt.ylabel("GDP Per Capita")
```

This line chart is much more descriptive in nature than the previous one. We can conclude that the United States has consistently better GDP per capita as compared to other countries in the list.

Let us now plot the data for US in a separate line chart. For this we will use ***loc***

```python
NorthA.loc["United States"].plot()
plt.ylabel("GDP per capita")
```

We can even plot line chart for different countries simulatneously.

```python
NorthA.loc["Cuba"].plot()
NorthA.loc["Jamaica"].plot()
NorthA.loc["Mexico"].plot()
```

Clearly, we are clueless as to which line plot belongs to which country. Hence, as a good coding hygiene one should always add legend to the plot to explain characters, symbols, or markings on a document that may be unfamiliar to the reader.

```python
NorthA.loc["Cuba"].plot()
NorthA.loc["Jamaica"].plot()
NorthA.loc["Mexico"].plot()
plt.legend()
```

#### Custom Line Color

To change the line color, use the **‘color’** argument.

```python
NorthA.loc["United States"].plot(color="red")
plt.ylabel("GDP Per Capita")
```

You can also add transparency to the color with the **'alpha'** argument (0=transparent, 1=opaque)

```python
NorthA.loc["United States"].plot(color='red', alpha=0.3)
NorthA.loc["Cuba"].plot(color='red', alpha=0.9)
plt.ylabel("GDP Per Capita")
```

#### Custom Line Style

You can choose between different line styles with the **'linestyle'** argument.
A few of the line styles available are:
- **-**
- **--**
- **-.**
- **:**


We will now see how these different line styles look like in a line plot.

```python
NorthA.loc['United States'].plot(color='red', alpha=0.8)
plt.ylabel("GDP Per Capita")
```

```python
NorthA.loc['United States'].plot(color='red', alpha=0.8, linestyle='--')
plt.ylabel("GDP Per Capita")
```

```python
NorthA.loc['United States'].plot(color='red', alpha=0.9, linestyle='-.')
plt.ylabel("GDP Per Capita")
```

```python
NorthA.loc['United States'].plot(color='red', alpha=0.5, linestyle=':')
plt.ylabel("GDP Per Capita")
```

#### Custom Line width

You can custom the line width usng the **'linewidth'** argument.

```python
NorthA.loc['United States'].plot(linewidth=5)
plt.ylabel("GDP Per Capita")
```

#### Add annotation in the line plot

```python
NorthA.loc['United States'].plot(color='red', linewidth=5, linestyle='-')
plt.ylabel("GDP Per Capita")
plt.annotate('GDP for USA', xy=(1967,30000), color='black')
```

Now we will add the observations on top of each point plotted in the line chart.
Consider the following example where we plot values generated from two populated lists.

```python
df = NorthA.loc['United States']
print(df)
df.plot()
```

```python
xs = list(years)
ys = [round(df.iloc[i]) for i in range(len(df))]
plt.plot(xs,ys,'bo-')
for x,y in zip(xs,ys):
    label = (x,y)
    plt.annotate(label, # this is the text
                 (x,y), # this is the point to label
                 textcoords="offset points", # how to position the text
                 xytext=(0,-6), # distance from text to points (x,y)
                 ha='right') # horizontal alignment can be left, right or center
```
